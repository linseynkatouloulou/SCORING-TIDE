Le fichier ou l'on peut retrouver les données avec les premieres analyses (c'est mieux de séparer si on veut jeter un coup d'oeil aux data au lieu de remonter à chaque fois)

• **Carte** = Le numero de carte

• **Date** = La date de la transaction

• **Montant** = Le montant de la transaction

• **Pays** = Le code du pays dans lequel la transaction a été réalisée

• **MCC** = Le code du commercant chez qui la transaction a été réalisée

• **CodeRep** = Code réponse de la demande d’autorisation : 00 = Accepté, les autres codes sont des refus

• **Fraude** = La variable fraude permet d’identifier les transactions frauduleuses

• **FM_Velocity_Condition** Nombre de transactions réalisées dans les X dernières heures pour chaque carte (Vélocité)

• **FM_sum** Montant cumulé des transactions réalisées dans les X dernières heures

• **FM_difference** Montant cumulé des transactions réalisées dans des pays différnents X dernières heures

• **FM_redondance** Nombre de transactiosn réalisées chez un même type de commerçant 




```python
import sys
import os

print("Current working directory:", os.getcwd())

sys.path.append(os.path.abspath('..'))
print("Updated sys.path:", sys.path)
from load_data import chargement_data

import pandas as pd
import zipfile
from sas7bdat import SAS7BDAT
```

```python
file_path = '../data/autorisations.sas7bdat'  

with SAS7BDAT(file_path) as reader:
    df = reader.to_data_frame()
```

```python
# Contrôle des valeurs manquantes
print(f'Nombre de valeurs manquantes : {df.count().isna().sum()}')
```

```python
# Tableau récapitulatifs sur les modalités de chaques variables

df_dtypes = df.dtypes.reset_index().rename(columns={'index':'Champ',0:'Type'})
df_dtypes

for i in df.columns:
    df_dtypes.loc[df_dtypes['Champ']==i,'Modalités'] = str(sorted(df[i].unique().tolist()))
    df_dtypes.loc[df_dtypes['Champ']==i,'NbDistinctVal'] = df[i].nunique()
    
df_dtypes
```

```python
# Dictionnaire de conversion des types
conversion_dict = {'Carte': 'object', 'Pays': 'object', 'Date': 'datetime64[ns]', 'Heure': 'datetime64[ns]', 'dateheure': 'datetime64[ns]',
                   'CodeRep': 'object', 'MCC': 'object', 'Montant': 'float64', 'fraude': 'object',
                   'FM_Velocity_Condition_3': 'int64', 'FM_Velocity_Condition_6': 'int64', 'FM_Velocity_Condition_12': 'int64',
                   'FM_Velocity_Condition_24': 'int64', 'FM_Sum_3': 'float64', 'FM_Sum_6': 'float64', 'FM_Sum_12': 'float64', 
                   'FM_Sum_24': 'float64', 'FM_Redondance_MCC_3': 'int64', 'FM_Redondance_MCC_6': 'int64', 
                   'FM_Redondance_MCC_12': 'int64', 'FM_Redondance_MCC_24': 'int64', 'FM_Difference_Pays_3': 'int64',
                   'FM_Difference_Pays_6': 'int64', 'FM_Difference_Pays_12': 'int64', 'FM_Difference_Pays_24': 'int64'}

# Application de la conversion
df = df.astype(conversion_dict)

df.dtypes
```

```python
df.columns = [col.lower() for col in df.columns]
df.head(10)
```

```python
# Tableau récapitulatifs sur les modalités de chaques variables

df_dtypes = df.dtypes.reset_index().rename(columns={'index':'Champ',0:'Type'})
df_dtypes

for i in df.columns:
    df_dtypes.loc[df_dtypes['Champ']==i,'Modalités'] = str(sorted(df[i].unique().tolist()))
    df_dtypes.loc[df_dtypes['Champ']==i,'NbDistinctVal'] = df[i].nunique()
    
df_dtypes
```
